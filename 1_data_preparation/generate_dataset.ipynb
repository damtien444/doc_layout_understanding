```python
from enum import Enum
from pathlib import Path


class CategoryId(Enum):
    ILLUSTRATION = 0
    TEXT = 1
    FORMULA = 2
    QUESTION = 3
    TITLE = 4
    EXPLANATION = 5
    ANSWER = 6
    SUB_QUESTION = 7
    SUPER_QUESTION = 20
    SUPER_TITLE = 21
    HEADER = 22
    FOOTER = 23
    HEADING = 24
    ENDING = 25

def convert_to_coco(result_path: str, output_name: str):
    assert result_path.endswith('.json')

    with open(result_path, 'r') as f:
        ocr_result = json.load(f)

    categories = []
    for cat in CategoryId:
        cat_name = cat.name.lower()
        categories.append({
            'id': cat.value,
            'name': cat_name,
            'supercategory': '',
        })

    coco_anns = []
    for ann in ocr_result['annotations']:
        x0, y0, x1, y1 = ann['bbox']
        segs = ann.get('segs', None)
        coco_anns.append({
            'id': ann['id'],
            'image_id': ann['image_id'],
            'category_id': ann['category_id'],
            'bbox': [x0, y0, x1 - x0, y1 - y0],
            'area': (x1 - x0) * (y1 - y0),
            'segmentation': segs if segs is not None else [],
            'iscrowd': 0,
            'attributes': {
                'value': ann['pred_values'],
                'score': ann['score'],
                'occluded': False,
            }
        })

    out_path = f"{Path(result_path).parent.as_posix()}/{output_name}.json"
    with open(out_path, 'w', encoding='utf-8') as g:
        json.dump({
            'categories': categories,
            'images': ocr_result['images'],
            'annotations': coco_anns,
        }, g, indent=4, ensure_ascii=False)

    return out_path
```

```python
from tqdm import tqdm


def _search(origin_instance, search_domain, comparative_feature):
    key = origin_instance[comparative_feature]
    for idx, instance in enumerate(search_domain):
        if instance['category_label']==3 or instance['category_label']==20 or instance['category_id']==1 or instance['category_id']==2 or instance['category_id']==3:
            continue
        if key == instance[comparative_feature]:
            return idx
    return -1

def _update(instance, features, values):
    for idx, feature in enumerate(features):
        instance[feature] = values[idx]
    return instance

def _get_values(instance, features):
    values = []
    for feature in features:
        values.append(instance[feature])
        
    return values

# complexity O(n^2)
def matching_and_flagging_counterpart(db1, search_domain, comparative_feature, update_features):
    for idx, instance in enumerate(tqdm(db1)):
        counter_idx = _search(instance, search_domain, comparative_feature)
        if counter_idx != -1:
            values = _get_values(search_domain[counter_idx], update_features)

            db1[idx] = _update(instance, update_features, values)
        else:
            pass
        
    return db1
```
